# Exploratory Data Analysis(EDA)

```python
# Univariate Analysis

cat=[]
con=[]

for i in A.columns:
    if(A[i].dtypes=="object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
for i in con:
    sb.distplot(A[i])
    plt.show()
```

```python
for i in cat:
    sb.countplot(A[i])
    plt.show()
```

```python
A["Type"].value_counts()
```

```python
A["Type"].value_counts().plot(kind="barh")
```

```python
A["Type"].value_counts().plot(kind="bar")
```

```python
A["Type"].value_counts().plot(kind="pie")
```

```python
# Bivarent Analysis
for i in A.columns:
    if(A[i].dtypes == "object"):
        sb.boxplot(A[i],A.Price)
        plt.xlabel(i)
        plt.ylabel("Price")
        plt.show()
    else:
        sb.scatterplot(A[i],A.Price)
        plt.xlabel(i)
        plt.ylabel("Price")
        plt.show()
        
```

```python
# Multivariate Analysis
sb.pairplot(A)
```
